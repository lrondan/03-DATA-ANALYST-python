```python
#regresion lineal simple MLR
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
#crear set de datos
d = {'carros':['MW', 'TOYOTA', 'MITSUBISHI', 'AURON', 'CADILAC', 'JDH', 'JASB', 'AHSBD','HASDV','AHDV'], 'precio':[1000,5461,1515,5156,5651,6515,6555,1565,2666,6956], 'puntos':[10,5,1,5,5,6,6,3,2,9], 'col4':[23,52,235,235,235,2,35,535,46,436], 'col5':[123,522,2435,2535,2135,21,315,5135,416,4136]}
df = pd.DataFrame(data=d)
print(df)
```

```python
#Entrenar modelo
lm = LinearRegression()
z = df[['col4','precio']]
lm.fit(z, df['puntos'])
#obtener prediccion
Yhat = lm.predict(z)
```

```python
#visualizar modelo
ax1 = sns.distplot(df['puntos'], hist=False, color='r', label='Puntos')
sns.distplot(Yhat, hist=False, color = 'b', label='Fitted Values', ax=ax1)
```
