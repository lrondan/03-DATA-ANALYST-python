```python
#regresion lineal simple SLR
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
#crear set de datos
d = {'carros':['MW', 'TOYOTA', 'MITSUBISHI', 'AURON', 'CADILAC', 'JDH', 'JASB', 'AHSBD','HASDV','AHDV'], 'precio':[1000,5461,1515,5156,5651,6515,6555,1565,2666,6956], 'puntos':[10,5,1,5,5,6,6,3,2,9]}
df = pd.DataFrame(data=d)
print(df)
```

```python
#extraer modelo lineal simple
X = df[['puntos']]
Y = df['precio']
#entrenar fit
lm = LinearRegression()
lm.fit(X, Y)
#hacer prediccion
Yhat = lm.predict(X)
```

```python
#acercamiento de la prediccion al modelo
ax1 = sns.distplot(df['precio'], hist=False, color='r', label='Precio')
sns.distplot(Yhat, hist=False, color = 'b', label='Fitted Values', ax=ax1)
```
