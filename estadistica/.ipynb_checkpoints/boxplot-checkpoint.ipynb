```python
import pandas as pd
import seaborn as sns
```

```python
d = {'drive':['A', 'B', 'C', 'D', 'E'], 'price':[120, 253, 145, 412, 146], 'drive weels':['wds', 'asd', 'fwa', 'fw4','l3']}
df = pd.DataFrame(data=d)
```

```python
dr_count = df['drive'].value_counts()
```

```python
dr_count.renamecolumns={'drive':'value_counts'}
dr_count.index.name='drive weels'
```

```python
#box plot
sns.boxplot(x = 'drive weels', y = 'price', data=df)
```
