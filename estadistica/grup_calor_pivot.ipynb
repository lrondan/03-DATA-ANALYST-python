```python
#importar libs
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#crear df
d = {'size':[78, 55, 51, 51], 'price':[23, 54, 65, 54], 'body-style':[0.2, 0.8, 0.3, .03]}
df = pd.DataFrame(data=d)
```

```python
#agrupar datos
df_test = df[['body-style', 'size', 'price']]
df_grup = df_test.groupby(['price'], as_index=True)
df_grup.head()
```

```python
a = df_grup.mean()
```

```python
print(type(a))
```

```python
#tabla de pivote
df_pivot = a.pivot(index='size', columns='body-style')
print(df_pivot)
```

```python
#mapa de calor
plt.pcolor(df_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
#mapa de calor
plt.pcolor(a, cmap='RdBu')
plt.colorbar()
plt.show()
```
