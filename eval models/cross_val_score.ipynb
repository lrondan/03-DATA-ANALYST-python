```python
#cross val score
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
#crear set de datos
d = {'carros':['MW', 'TOYOTA', 'MITSUBISHI', 'AURON', 'CADILAC', 'JDH', 'JASB', 'AHSBD','HASDV','AHDV'], 'precio':[1000,5461,1515,5156,5651,6515,6555,1565,2666,6956], 'puntos':[10,5,1,5,5,6,6,3,2,9], 'col4':[23,52,235,235,235,2,35,535,46,436], 'col5':[123,522,2435,2535,2135,21,315,5135,416,4136]}
df = pd.DataFrame(data=d)
print(df)
```

```python
#entrenar datos
lr = LinearRegression()
x_values = df[['puntos']]
y_values = df['precio']

lr.fit(x_values, y_values)
```

```python
#validacion del modelo aunque no esta entrenado
scores = cross_val_score(lr, x_values, y_values, cv=3)
results=np.mean(scores)
print(f'la mediana de los valores de R-score del set de datos es {results}')
```
