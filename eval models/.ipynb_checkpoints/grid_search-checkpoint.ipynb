```python
#regresion lineal simple MLR
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import GridSearchCV
import seaborn as sns
```

```python
#crear set de datos
d = {'carros':['MW', 'TOYOTA', 'MITSUBISHI', 'AURON', 'CADILAC', 'JDH', 'JASB', 'AHSBD','HASDV','AHDV'], 'precio':[1000,5461,1515,5156,5651,6515,6555,1565,2666,6956], 'puntos':[10,5,1,5,5,6,6,3,2,9], 'col4':[23,52,235,235,235,2,35,535,46,436], 'col5':[123,522,2435,2535,2135,21,315,5135,416,4136]}
df = pd.DataFrame(data=d)
#Entrenar modelo
lm = LinearRegression()
x = df[['col4']]
y = df['precio']
lm.fit(x,y)
#obtener prediccion
Yhat = lm.predict(x)
```

```python
parametros = [{'alpha': [0.0001, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000, 10000000]}]
```

```python
RR=Ridge()
```

```python
Grid1 = GridSearchCV(RR, parametros, cv=4)
```

```python
Grid1.fit(x,y)
```

```python
Grid1.best_estimator_
```

```python
scores = Grid1.cv_results_
ms =scores['mean_test_score']
split0=scores['split0_test_score']
split1=scores['split1_test_score']
split2=scores['split2_test_score']
split3=scores['split3_test_score']
```

```python
#visualizar modelo
sns.distplot(split0, hist=False, color='b', label='split0')
sns.distplot(split1, hist=False, color='b', label='split1')
sns.distplot(split2, hist=False, color='b', label='split2')
sns.distplot(split3, hist=False, color='b', label='split3')
sns.distplot(ms, hist=False, color = 'r', label='Fitted Values')
```
