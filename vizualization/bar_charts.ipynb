```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import openpyxl
df = pd.read_excel('data/canadian_immegration_data.xlsx')
```

```python
years = list(map(str, range(1980, 2014)))
df_top5 = df.head()
df_top5 = df_top5[years].transpose()
```

```python
df_afg = df_top5[0]
```

```python
df_afg.plot(kind='bar')
plt.title('emigrantes afganist...')
plt.ylabel('total emigrantes')
plt.xlabel('anos')
plt.show()
```

```python
df_afg.plot(kind='barh')
plt.title('emigrantes afganist...')
plt.xlabel('total emigrantes')
plt.ylabel('anos')
plt.show()
```
